```python
import tweepy
import twitter
import json
import datetime
```

```python
consumer_key = ""
consumer_secret = ""
access_token = ""
access_token_secret = ""
```

```python
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth,wait_on_rate_limit=True)
```

```python
# Extract tweets from January 1, 2020 from Trump
trump = {}
all_posts = api.user_timeline(screen_name="joebiden",count = 3200, lang ="en", tweet_mode="extended")

# basic information on individual

# initializing each tweet object within dictionary
# i = 1
# for tweet in posts:
#     jsonf = tweet._json
#     tweet_num = "tweet" + str(i)
#     trump[tweet_num] = {}
#     i = i + 1

trump["tweets"]={}
i = 1
#maxid="1317273920869928961"
```

```python
#for i in range(1,20):
posts = []
fromDate = datetime.datetime(2020,1,1,0,0)
for tweets in all_posts:
    date_raw = tweets._json["created_at"]
    date_new = datetime.datetime.strptime(date_raw,'%a %b %d %H:%M:%S +0000 %Y')
    if date_new > fromDate:
        posts.append(tweets)
maxid = tweets.id
# extracting desired information and storing in dictionary
for tweet in posts[:1]:
    jsonf = tweet._json
    trump["name"] = jsonf["user"]["name"]
    trump["followers"] = jsonf["user"]["followers_count"]
    trump["friends"] = jsonf["user"]["friends_count"]
    trump["total_tweets"] = jsonf["user"]["statuses_count"]

for tweet in posts:
    jsonf = tweet._json
    tweet_num = str(i)

    # date of tweet
    trump["tweets"][tweet_num]={}
    trump["tweets"][tweet_num]["date"] = jsonf["created_at"]

    # full text of tweet
    trump["tweets"][tweet_num]["text"] = jsonf["full_text"]

    # number of views (not unique to user)
    # trump[tweet_num]["views"] = tweet.impression_count

    # retweets count
    trump["tweets"][tweet_num]["retweets"] = jsonf["retweet_count"]

    # favorites count
    trump["tweets"][tweet_num]["favorites"] = jsonf["favorite_count"]

    # replies count
    # trump[tweet_num]["views"] = tweet.reply_count

    # hashtags
    trump["tweets"][tweet_num]["hashtags"] = []
    if jsonf["entities"]["hashtags"] == []:
        trump["tweets"][tweet_num]["hashtags"] = []
    else:
        lst = jsonf["entities"]["hashtags"]
        for item in lst:
            trump["tweets"][tweet_num]["hashtags"].append(item["text"])

    # media (photo, video in tweet)
    trump["tweets"][tweet_num]["media"] = ""
    if "media" in jsonf["entities"]:
        lst = jsonf["entities"]["media"]
        for item in lst:
            trump["tweets"][tweet_num]["media"] = item["type"]
    else:
        trump["tweets"][tweet_num]["media"] = ""
    i = i + 1

#print(json.dumps(trump, indent=4))
```
